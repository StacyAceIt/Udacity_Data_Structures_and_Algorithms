Problem statement
Given an array arr and a target element target, find the last index of occurence of target in arr using recursion. If target is not present in arr, return -1.

For example:

For arr = [1, 2, 5, 5, 1, 2, 5, 4] and target = 5, output = 6

For arr = [1, 2, 5, 5, 1, 2, 5, 4] and target = 7, output = -1

```python
def lastIndex(arr, target):
  return lastIndex_(arr, target, len(arr)-1)

def lastIndex_(arr, target, index):
  # print("arr {}".format(arr))
  # print("index {}".format(index))
  # print("arr[index] = {}".format(arr[index]))
  # print("target = {}".format(target))
  if index < 0:
    print("Not found")
    return -1

  if arr[index] == target:
    print("Found: index = {}".format(index))
    return index

  return lastIndex_(arr, target, index-1)

print(lastIndex([1, 2, 5, 5, 1, 2, 5, 4],5))
print(lastIndex([1, 2, 5, 5, 1, 2, 5, 4],7))
```
